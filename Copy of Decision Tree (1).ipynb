```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Purchasenew = pd.read_csv('Purchase_new.csv')
print(Purchasenew)
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
```

```python
Purchasenew['Holiday'] = Le.fit_transform(Purchasenew['Holiday'])
Purchasenew['Discount'] = Le.fit_transform(Purchasenew['Discount'])
Purchasenew['Free Delivery'] = Le.fit_transform(Purchasenew['Free Delivery'])

Purchasenew['Purchase'] = Le.fit_transform(Purchasenew['Purchase'])
```

```python
y = Purchasenew['Purchase']
X = Purchasenew.drop(['Purchase'],axis=1)
```

```python
# Fitting the model
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion = 'entropy')
clf = clf.fit(X, y)
```

```python
# We can visualize the tree using tree.plot_tree
tree.plot_tree(clf)
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph
```
